# Project 18: Sort a List

```python
# This project teaches how to sort numbers in ascending order
```

```python
# Step 1: Ask the user to enter numbers separated by commas
numbers_input = input("Enter numbers separated by commas: ")
```

```python
# Step 2: Convert the string into a list of numbers
numbers_list = [float(x) for x in numbers_input.split(",")]
```

```python
# Step 3: Sort the list
numbers_list.sort()  # sorts the list in ascending order
```

```python
# Step 4: Print the sorted list
print(f"The sorted list is: {numbers_list}")
```
